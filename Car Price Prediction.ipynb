# Data Cleaning

```python
#name column - keep first 3 words
df['Name']=df['Name'].str.split(' ').str.slice(0,3).str.join(' ')
df['Name']=df['Name'].str.replace('-','')
df['Name']
```

```python
#Price - remove ask for price
c=0
for i in df['Price']:
    if i == 'Ask For Price':
        c+=1
print(c)
```

```python
df=df[df['Price']!='Ask For Price']
df['Price']
```

```python
#remove special characters
df['Price']=df['Price'].str[1:].str.replace(',','').astype(int)
df['Price']
```

```python
#Kms_driven
df['Kms_driven']=df['Kms_driven'].str.split(' ').str[0].str.replace(',','').astype(int)
df['Kms_driven']
```

```python
df.info()
```

```python
#Fuel type
df['Fuel_type']=df['Fuel_type'].str.replace(' ','')
df['Fuel_type'].value_counts()
```

```python
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(columns='Price')
y=df['Price']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.23,random_state=100)
```

```python
cols=['Name','Fuel_type','Owner','Company','Year']
ohe=OneHotEncoder()
ohe.fit(df[cols])
```

```python
ohe.categories_
```

```python
cols=['Name','Fuel_type','Owner','Company','Year']
col_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),cols),remainder='passthrough')
lr=LinearRegression()
pipe=make_pipeline(col_trans,lr)
pipe.fit(x_test,y_test)
```

```python
r2_score(y_test,pipe.predict(x_test))
```
